```python


def ricker(A,f,prof,RHOsi,VP):

#---------impedancia---------#

    z_imp = []

    for i in range (len(VP)):
        z_imp0 = (RHOsi[i]*VP[i])
        z_imp.append(z_imp0)
    
#-------refletividade--------#

    R  = []

    for i in range (len(VP)):
        if i == 0:
            RR = 0
        if i != 0:
            RR = (z_imp[i-1] - z_imp[i])/(z_imp[i-1] + z_imp[i])
        R.append(RR)
    
    zero_bar = [0]*len(VP)

#-----------tempo------------#


    DL = prof[3]-prof[2]# valor do intervalo em m

    T  = [] #tempo por intervalo padrao

    for i in range (len(VP)):
        T.append(DL/VP[i])

#----------------------------#

    mp  = (len(VP) - 1)/2 # termo central

    dtf = [0.0]*mp
    dtb = [0.0]*mp

    for i in range (mp):
        dtf[i] = np.mean(T)
        dtb[i] = -(np.mean(T))
    
    for i in range (1,mp):
        dtf[i]  = dtf[i- 1] + dtf[i]
    
    for i in range (1,mp):
        dtb[i]  = dtb[i -1] + dtb[i]

    dtb.reverse()

#----------------------------#

    TT  = [0.0]*p


    for i in range (len(dtb)):
        TT[i]  = dtb[i]

    for i in range (mp+1,p):
        TT[i]  = dtf[i-mp-1]
    
#----------wavlet------------#

    ff = str(f) + " - Hz"
    j = 400

    wav = []
    for i in range (len(TT)):
        EE1 = -(np.pi**2)*(f**2)*(TT[i]**2)
        wav01 = A*(1 - (2*(np.pi**2)) * (f**2) * (TT[i]**2))*np.exp(EE1) # 2497.55919007
        wav.append(wav01)
    
    signal = np.convolve(R,wav,'same')  # wavlet convolucao Padrao
    
    resultados = [z_imp,R,TT,wav,signal]

    return(resultados)
```
